# 1. Apply Convolution using both smoothing and sharpening kernels on a grayscale image and output the results

- Gaussian Smoothing Filter



## a. Smoothing kernel is a gaussian filter of size 5 x sigma  

```python
sm_kernel = gaussian_smoothing_kernel(15, 1.0)

sm_convo = apply_convolution(lena, sm_kernel)
```

```python
padding = sm_convo["padding"]

lena_border = sm_convo["border_img"]

raw_conv = sm_convo["raw_result"]

lena_norm = sm_convo["normalized"]

lena_norm_cropped = sm_convo["final"]
```

```python
# Show all images
cv2.imshow('Original Grayscale Image', lena)
cv2.imshow('Bordered Image', lena_border)
cv2.imshow('Convolution Image', raw_conv)
cv2.imshow('Normalized Image', lena_norm)
cv2.imshow('Normalized Cropped Image', lena_norm_cropped)


cv2.waitKey(0)
cv2.destroyAllWindows()
```
